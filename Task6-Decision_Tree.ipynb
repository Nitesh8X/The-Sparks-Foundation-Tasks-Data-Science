For the given 'Iris' dataset, we created the Decision Tree classifier and visualized it graphically.
The purpose of this task is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
#import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import confusion_matrix
```

```python
iris = pd.read_csv('irisdata.csv')
iris.head()
```

```python
iris.shape
```

```python
iris.describe()
```

## Data Visualization

```python
sns.pairplot(iris, hue = 'Species')
plt.show()
```

Here we see class 'setosa' has a distinct cluster. 

```python
#categorical features of different species

plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x="Species", y="Sepal.Length", data=iris)
plt.subplot(2,2,2)
sns.boxplot(x="Species", y="Sepal.Width", data=iris)
plt.subplot(2,2,3)
sns.boxplot(x="Species", y="Petal.Length", data=iris)
plt.subplot(2,2,4)
sns.boxplot(x="Species", y="Petal.Width", data=iris)
```

```python
#correlation

plt.figure(figsize=(10, 11))
sns.heatmap(iris.corr(),annot=True)
plt.plot()
```

We got see that Petal length is related to petal width.

## Decision Tree Model

```python
feature_cols = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
X = iris[feature_cols]
y = iris.Species

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
print("Trained sucessfully")

y_pred=dtc.predict(X_test)
```

## Accuracy of Model

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(dtc,filled=True, rounded=True, proportion=True, node_ids=True)
plt.show()
```
